Difference between iloc and loc
ILOC:-

The iloc is a integer-location function that is based on indexing for selection by position.
.iloc[] is primarily integer position based (from 0 to length-1 of the axis).inputs that are accepted are:
* an integer,
* a list or array of integers,
*a slice object with ints,
*boolean arrays.

LOC:-
The loc access a group of rows and columns by label(s) or a boolean array.

.loc[] is primarily label based,
the accepted Input are:

* a list or array,

* single label

* a boolean array of the same length being sliced


[Problem 6] Creating a diagram

Pie chart of the number of samples per label (while showing percentages)

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
labels = data.target_names
explode = (0.1, 0.1, 0.1)
sizes =  [len(df.loc[df["Species"]]== percentage) for percentage in range(len(data.target_names))]
fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, explode=explode,colors=['pink', 'brown', 'saddlebrown'], autopct='%1.1f%%',pctdistance=1.25, labeldistance=.6,shadow=True, startangle=120)
plt.title('Distribution of Samples per IRIS Plants')
plt.show()
```

* Selecting the features one by one and visualizing the distribution of data for each label using a box plot

```python
features =  ["sepal_length", "sepal_width", "petal_length", "petal_width"]
species = data.target_names

for feature in features:
    a_list = []
    for a in range(len(species)):
        species_found = df.loc[df["Species"] == a]
        a_list.append(species_found[feature])

    fig, ax = plt.subplots()
    ax.set_title(feature)
    ax.boxplot(a_list)
```

BOX PLOT VS VIOLIN PLOT

BOX PLOT

Boxplots and Violin plots are both useful tools for visualizing the distribution of data and comparing different groups.
 A boxplot provides a summary of the distribution of a dataset by displaying key percentiles (e.g., quartiles) and identifying outliers.
 Advantages:
 * Provides a clear summary of the distribution.
 * Compact and easy to interpret.
 * Good for comparing multiple groups side by side.

VIOLIN PLOT

A Violin plot combines aspects of a boxplot and a kernel density plot to provide a more detailed representation of the data distribution.
Advantages:
* Shows the shape of the distribution more explicitly.
* Helpful for identifying bimodal or skewed distributions.
* Provides a good compromise between summary statistics and detailed distribution
