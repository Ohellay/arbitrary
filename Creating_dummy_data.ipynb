[Problem 1] Creating random numbers

```python
import numpy as np
np.random.seed(0)
# Creating the Mean Vector
mean = [-3, 0]
# Here is the covariance
covariance =np.array([[1.0, 0.8], [0.8,1]] )
#The dummy_data store the generated random data (bivariate normal distribution)
dummy_data = np.random.multivariate_normal(mean, covariance, 500)
print(dummy_data)

shape = np.shape(dummy_data)
print(shape)
```

[Problem 2] Visualization by scatter plot

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)

mean = [-3, 0]
covariance = np.array([[1.0, 0.8], [0.8, 1]])
dummy_data = np.random.multivariate_normal(mean, covariance, 500)

# Plotting Scatter Plots
plt.scatter(dummy_data[:, 0], dummy_data[:, 1], alpha=0.5)
plt.title('Scatter Plot of Bivariate Normal Distribution')
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```

Visualize the data created in Problem 1 with a histogram.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)


mean = [-3, 0]

covariance = np.array([[1.0, 0.8], [0.8, 1]])
dummy_data = np.random.multivariate_normal(mean, covariance, 500)

# Plotting histograms
plt.figure(figsize=(10, 4))

# Histogram for Variable 1
plt.subplot(1, 2, 1)
plt.hist(dummy_data[:, 0], bins=30, color='red', alpha=0.5)
plt.title('Histogram of Variable 1')
plt.xlabel('x1')
plt.ylabel('Frequency')

# Histogram for Variable 2
plt.subplot(1, 2, 2)
plt.hist(dummy_data[:, 1], bins=30, color='black', alpha=0.5)
plt.title('Histogram of Variable 2')
plt.xlabel('x2')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

PROBLEM 4

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

np.random.seed(0)

mean = [0,-3]
covariance =np.array([[1.0, 0.8], [0.8,1]] )


dummy_data_new = np.random.multivariate_normal(mean, covariance, 500)


plt.scatter(dummy_data[:, 0], dummy_data[:, 1], alpha=0.5, label='0')
plt.scatter(dummy_data_new[:, 0], dummy_data_new[:, 1], alpha=0.5, label='1')
plt.title('Scatter Plot of Bivariate Normal Distributions')
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend()
plt.show()
```

[Problem 5] Data combination

>  np.concatenate() Method



```python
import numpy as np



np.random.seed(0)
# The  concatenation Method
concat = np.concatenate((dummy_data, dummy_data_new), axis=0)
print(concat)
shape = np.shape(concat)
print(shape)
```

VSTACK Method

```python
import numpy as np

np.random.seed(0)

# The VStack method
vstack = np.vstack((dummy_data,dummy_data_new))
print(vstack)
shape = np.shape(vstack)
print(shape)
```
