PROBLEM 4

COMPARING BOTH CODES


*This is the First Instance*

```python
%%timeit
import time
start = time.time()

THICKNESS = 0.00008

fold_time = 43

folded_thickness = THICKNESS * 2 ** fold_time

print("Thickness: {} meters".format(folded_thickness))
elapsed_time = time.time() - start
print("time : {}[s]".format(elapsed_time))
```

This is the second Instance

```python
%%timeit
import time
start = time.time()

THICKNESS = 0.00008
fold_time = int(43)

for i in range(1,fold_time +1 ) :
  folded_thickness = THICKNESS * 2 ** i

print("Thickness: {} meters".format(folded_thickness))

elapsed_time = time.time() - start
print("time : {}[s]".format(elapsed_time))
```

PROBLEM 5

This records the total of 44 process values to the code and confirm that 44 values are stored in the list.


```python
folded_thickness_list = []
THICKNESS = 0.00008
fold_time = int(44)

for i in range(1,fold_time +1 ) :
  folded_thickness = THICKNESS * 2 ** i
  folded_thickness_list.append(folded_thickness)
  print("The list are", folded_thickness_list)
length = len(folded_thickness_list)
print("The length of the folded_thickness_list::", length)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

# Display the graph. Title and axis label name.
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.plot(folded_thickness_list, color="yellow")
plt.tick_params(labelsize=20) # Make settings related to axis values
plt.show()
```
