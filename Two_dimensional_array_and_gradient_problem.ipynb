[Problem 1] Linear function

````python
import numpy as np

#creating the X-array
x = np.arange(-50, 50.1, 0.1)

print(x)
y = 0.5 * x + 1
print(y)

x.shape
````

[Problem 2] Array combination

````python
import numpy as np

col1, col2 = x, y
# Creating a new array using np.array
new_array = np.vstack((col1, col2)).T
# or using new_array = np.column_stack((col1, col2))

print(new_array)
# Printing the shape of the new array
print(new_array.shape)
````

[Problem 3] Find the gradient


````python
# To find the Gradient
gradient = (new_array[1: ,1] - new_array[:-1,1])/(new_array[1:,0]-new_array[:-1,0])
print(gradient)


#note: gradient = np.diff(new_array[:, 1]) / np.diff(new_array[:, 0]) can also be used, It find the difference between the values.
````

[Problem 4] Draw a graph

> Draw a graph of the linear function using Matplotlib. Also, draw a graph with x on the horizontal axis and gradient on the vertical axis.



````python
import matplotlib.pyplot as plt
%matplotlib inline


# Plotting the linear function
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(new_array[:,0], new_array[:,1], label='Linear Function: 0.5 * x + 1')
plt.title('Linear Function')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

# Plotting the gradient
plt.subplot(1, 2, 2)
plt.plot(new_array[:-1, 0], gradient, label='Gradient')
plt.title('Gradient')
plt.xlabel('x')
plt.ylabel('Gradient')
plt.legend()

plt.tight_layout()
plt.show()
````

[Problem 5] Python functionalization

````python
def compute_gradient(function, x_range=(-50, 50.1, 0.1)):
    """
    Calculate the gradient using the amount of change.

    Parameters
    ----------------
    function : function
      The function you want to find, the one that returns the ndarray of y when you put the ndarray of x.
    x_range : tuple
      Specify the range in the same way as the argument of np.arange ().

    Returns
    ----------------
    array_xy : ndarray, shape(n, 2)
      A combination of x and y. n depends on x_range.
    gradient : ndarray, shape(n-1,)
      Function gradient. One factor is reduced to make a difference
    """
    array_x = np.arange(*x_range)
    array_y = function(array_x)
    array_xy =  np.vstack((array_x, array_y)).T
    gradient = (array_xy[1: ,1] - array_xy[:-1,1])/(array_xy[1:,0]-array_xy[:-1,0])
    return array_xy, gradient
def function1(array_x):
  array_y = array_x**2
  return array_y

array_xy1, gradient1 = compute_gradient(function1)
print(compute_gradient(function1))
````

````python
def compute_gradient(function, x_range=(0, 50.1, 0.1)):
    """
    Calculate the gradient using the amount of change.

    Parameters
    ----------------
    function : function
      The function you want to find, the one that returns the ndarray of y when you put the ndarray of x.
    x_range : tuple
      Specify the range in the same way as the argument of np.arange ().

    Returns
    ----------------
    array_xy : ndarray, shape(n, 2)
      A combination of x and y. n depends on x_range.
    gradient : ndarray, shape(n-1,)
      Function gradient. One factor is reduced to make a difference
    """
    array_x = np.arange(*x_range)
    array_y = function(array_x)
    array_xy =  np.vstack((array_x, array_y)).T
    gradient = (array_xy[1: ,1] - array_xy[:-1,1])/(array_xy[1:,0]-array_xy[:-1,0])
    return array_xy, gradient
def function3(array_x):
  array_y = np.sin(array_x**(0.5))
  return array_y

array_xy3, gradient3 = compute_gradient(function3)
print(compute_gradient(function3))
````
